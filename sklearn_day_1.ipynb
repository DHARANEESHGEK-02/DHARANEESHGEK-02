```python
import sklearn
```

```python
from sklearn.datasets import load_iris
```

```python
load_iris()
```

```python
load_iris(return_X_y =True)
```

```python
X,y=load_iris(return_X_y=True)
```

```python
from sklearn.linear_model import LogisticRegression as lr
```

```python
model=lr()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
from sklearn.neighbors import RadiusNeighborsRegressor as nr
```

```python
mod=nr()
mod.fit(X,y)
```

```python
mod.predict(X)
```

```python
import matplotlib.pyplot as plt
pred=model.predict(X)
plt.scatter(pred,y)
```

```python
import pandas as pd
```

```python
from sklearn.datasets import fetch_openml
df=fetch_openml('titanic',version=1,as_frame=True)['data']
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
sns.set()
mis_val=pd.DataFrame(df.isnull().sum()/len(df)*100)
mis_val.plot(kind='bar',title='Missing values in percentage',ylabel='Percentage')
```
